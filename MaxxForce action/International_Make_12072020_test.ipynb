```python
import requests
import json
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\FFR0103\Desktop\My files\WORK\Auto part\Maxxforce action\INTERNATIONAL make 12072020.xlsx',sheet_name='Report')
```

```python
test = df[(df['ValidationYear']>=2010) & (df['ValidationYear']<=2015)]
test
```

```python
responses = pd.Series(None, index=range(len(test)))
responses
```

```python
for i in range(len(test)):
    print(i)
    VIN = test.iloc[i,3]
    r = requests.post('https://secure.maxxforce11and13.com/en/VinLookup',{'VinNumber':VIN})
    soup = BeautifulSoup(r.text,'lxml')
    target = soup.find('form').find('div',class_ = 'alert alert-danger')
    if target:
        result= re.search(r'( {32})([\w\s]*\.)',str(target))
        responses[i] = result.group(2)
    else:
        target = soup.find('form').find('div',class_ = 'alert alert-success')
        result= re.search(r'( {32})([\w\s]*\.)',str(target))
        responses[i] = result.group(2)
responses
```

```python
test.insert(4,'response',responses.values)
```
