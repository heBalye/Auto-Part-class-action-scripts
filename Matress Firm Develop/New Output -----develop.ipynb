```python
_ = r'C:\Users\FFR0103\Desktop\My files\WORK\Auto part\Mattress Firm\Source\\'
Dict = {}
for file in header_dict:
    print(file)
    for i,sheet in enumerate(header_dict[file]):
        if header_dict[file][sheet] == 'NO VIN or State':
            df = pd.DataFrame({'FileName' :file, 'Index of Sheet':i+1,'SheetName': [str(sheet)],'row Contains Information':['NO']})
        elif header_dict[file][sheet] == 'Empty Sheet':
            df = pd.DataFrame({'FileName' :file, 'Index of Sheet':i+1,'SheetName': [str(sheet)],'row Contains Information':['Empty Sheet']})
        else:
            header = int(header_dict[file][sheet]) - 1
            df = pd.read_excel(_ + file,sheet_name=sheet,header = header)
            df = df.loc[:,df.columns.str.contains('state|vin|date|Auto ID Card Sent|In Service|pis|detailed',regex = True, case = False,na=False)]
            df.insert(0,'FileName',file)
            df.insert(1,'Index of Sheet',i+1)
            df.insert(2,'SheetName' ,sheet)
            df.insert(3,'row Contains Information','Yes')
            df.insert(4,'headerrow', header)
            df.insert(5,'number of record',len(df))
        if i == 0:
            Dict[file] = df
        else:
            Dict[file] = pd.concat([Dict[file],df],sort=False)
            
            

            
```

```python
result = pd.concat(Dict,sort=False)
result.to_excel('C:\\Users\\FFR0103\\Desktop\\My files\\WORK\\Auto part\\Mattress Firm\\Output\\Result_with_date.xlsx')
```
