## QC Table

```python
Dict = {}
for i,file in enumerate(FileName):
    print('#of file',i)
    DF = pd.read_excel(file,sheet_name=None)
    Dict[file.split('\\')[-1]] =pd.DataFrame()
    for i,sheet in enumerate(DF):

        if len(DF[sheet]) > 0:
            if DF[sheet].columns.str.contains('state|vin',regex = True, case = False,na=False).sum() > 0:
                df = pd.DataFrame({'File Name' :file.split('\\')[-1], 'Index of Sheet':i+1,'SheetName': [str(sheet)],'Firstrow Contains Information':['Yes']})
                
            else:
                df = pd.DataFrame({'File Name' :file.split('\\')[-1], 'Index of Sheet':i+1,'SheetName': [str(sheet)],'Firstrow Contains Information':['No']})
        else:
            df = pd.DataFrame({'File Name' :file.split('\\')[-1],'Index of Sheet':i+1,'SheetName': [str(sheet)],'Firstrow Contains Information':['Empty Sheet']})
        Dict[file.split('\\')[-1]] = pd.concat([Dict[file.split('\\')[-1]],df],sort=False)
QC_table = pd.concat(Dict,sort = False)
QC_table
```

## Skip row Features

```python
FileName = glob.glob('C:\\Users\\FFR0103\\Desktop\\My files\\WORK\\Auto part\\Mattress Firm\\Source\\*.*')
display(len(FileName))
Dict = {}
for skiprows in range(25):
    print(skiprows)
    for i,file in enumerate(FileName):
        DF = pd.read_excel(file,sheet_name=None,skiprows = skiprows, dtype=str)
        for sheet in DF:
            if file.split('\\')[-1]  not in Dict:
                if len(DF[sheet]) > 0:
                    if DF[sheet].columns.str.contains('state|vin',regex = True, case = False,na=False).sum() > 0:
                        df = DF[sheet].loc[:,DF[sheet].columns.str.contains('state|vin',regex = True, case = False,na=False)]
                        df.insert(0,'FileName',file.split('\\')[-1])
                        df.insert(1,'Sheet' ,sheet)
                        df.insert(2,'headerrow', skiprows)
                        df.insert(3,'number of record',len(DF[sheet]))
                        Dict[file.split('\\')[-1]] = df
            elif sheet not in Dict[file.split('\\')[-1]].Sheet.values:
                 if len(DF[sheet]) > 0:
                    if DF[sheet].columns.str.contains('state|vin',regex = True, case = False,na=False).sum() > 0:
                        df = DF[sheet].loc[:,DF[sheet].columns.str.contains('state|vin',regex = True, case = False,na=False)]
                        df.insert(0,'FileName',file.split('\\')[-1])
                        df.insert(1,'Sheet' ,sheet)
                        df.insert(2,'headerrow', skiprows)
                        df.insert(3,'number of record',len(DF[sheet]))
                        Dict[file.split('\\')[-1]] = pd.concat([Dict[file.split('\\')[-1]],df],sort=False)
                
                
                

result = pd.concat(Dict,sort=False)
result
result.to_excel('C:\\Users\\FFR0103\\Desktop\\My files\\WORK\\Auto part\\Mattress Firm\\Output\\result_skiprow25.xlsx')
```

## Clean Up the Output 

```python
import numpy as np
df = pd.read_excel('C:\\Users\\FFR0103\\Desktop\\My files\\WORK\\Auto part\\Mattress Firm\\Output\\01062020\\QC_result_formatted_01062020.xlsx', sheet_name='Details', index_col = [0,1])


StandardVIN = pd.Series(None,index = range(len(df)))

StandardState = pd.Series(None,index = range(len(df)))
```

### get VIN 

```python
df['# of  Value from selected VIN columns']=6-df[df.columns[[4,6,7,12,13,15]]].isnull().sum(axis = 1)

display(df['# of  Value from selected VIN columns'].unique())

for i in range(len(df)):
    temp = []
    for val in df.iloc[i,[4,6,7,12,13,15]]:
        if not pd.isnull(val):
            temp.append(str(val))
            StandardVIN[i] = ';'.join(temp)
StandardVIN 
```

### get State 

```python
df['# of  values from selected State columns']=6-df[df.columns[[5,8,9,11,16,17]]].isnull().sum(axis = 1)
display(df['# of  values from selected State columns'].unique())
for i in range(len(df)):
    temp = []
    for val in df.iloc[i,[5,8,9,11,16,17]]:
        if not pd.isnull(val):
            temp.append(str(val))
            StandardState[i] = ';'.join(temp)
            
StandardState
```

```python
df['StandardVIN'] = StandardVIN.values
df['StandardState'] = StandardState.values
df.to_excel('C:\\Users\\FFR0103\\Desktop\\My files\\WORK\\Auto part\\Mattress Firm\\Output\\CleanOutput.xlsx',index = False)
```
